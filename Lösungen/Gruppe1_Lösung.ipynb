# Gruppenarbeit 1 (Collider Bias)

#### Gegeben ist ein DataFrame, der Daten über die Körpergröße ("height") und der Geschwindigkeit ("speed") von Basketballspielern enthält. Spieler der Basketball-Profiliga NBA werden in der Spalte "nba" mit True gekennzeichnet.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Gruppe1_Daten.csv", index_col=False).drop("Unnamed: 0",axis=1)
df.head()
```

```python
df.columns
```

Aufgabe 1: Berechne die Korrelation für den NBA Datensatz. Zeichne einen Scatterplot. Was kannst du in den Daten erkennen?

```python
data_nba = df[df["NBA Spieler"] == True].copy()
```

    Lösung:

```python
corr_matrix = data_nba.corr()
corr = corr_matrix.loc['Körpergröße', 'Geschwindigkeit']
print(f"Korrelation für NBA Spieler: {corr}")

# Plot for NBA data
sns.scatterplot(data=data_nba, x="Körpergröße", y="Geschwindigkeit")
plt.xlabel("Körpergröße")
plt.ylabel("Geschwindigkeit")
plt.title("NBA data")

plt.show()
```

Aufgabe 2: Berechne die Korrelation für den gesamten Datensatz. Zeichne einen Scatterplot. Was kannst du in den Daten erkennen? Um welchen der vorgestellten Themen handelt es sich? 

```python
data = df.copy()
```

    Lösung:

```python
corr_matrix = data.corr()
corr = corr_matrix.loc['Körpergröße', 'Geschwindigkeit']
print(f"Korrelation (gesamt): {corr}")

# Plot for all data
sns.scatterplot(data=data, x="Körpergröße", y="Geschwindigkeit", hue="NBA Spieler")
plt.xlabel("Körpergröße")
plt.ylabel("Geschwindigkeit")
plt.title("All data")

plt.show()
```
